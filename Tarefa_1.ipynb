```python
import os
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
```

```python
train_df = pd.read_csv(os.path.join('Dados', 'train.csv'))
test_df = pd.read_csv(os.path.join('Dados', 'test.csv'))
```

```python
train_df.head()
```

```python
train_df.describe()
```

```python
train_df.target.unique()
```

```python
X = train_df.drop(columns=['id', 'target'])
y = train_df.target
```

### Regressão Logística

```python
reg_log = LogisticRegression()
```

```python
cv_results = cross_validate(reg_log, X, y, cv=10, scoring=('accuracy', 'f1_macro'))
```

```python
print('TREINO')
print('Acc: %.2f (+/- %.2f)' % (np.average(cv_results['train_accuracy']), np.std(cv_results['train_accuracy'])))
print('F1: %.2f (+/- %.2f)' % (np.average(cv_results['train_f1_macro']), np.std(cv_results['train_f1_macro'])))

print('TESTE')
print('Acc: %.2f (+/- %.2f)' % (np.average(cv_results['test_accuracy']), np.std(cv_results['test_accuracy'])))
print('F1: %.2f (+/- %.2f)' % (np.average(cv_results['test_f1_macro']), np.std(cv_results['test_f1_macro'])))
```

### Naive Bayes

```python
nb = GaussianNB()
```

```python
cv_results = cross_validate(nb, X, y, cv=10, scoring=('accuracy', 'f1_macro'))
```

```python
print('TREINO')
print('Acc: %.2f (+/- %.2f)' % (np.average(cv_results['train_accuracy']), np.std(cv_results['train_accuracy'])))
print('F1: %.2f (+/- %.2f)' % (np.average(cv_results['train_f1_macro']), np.std(cv_results['train_f1_macro'])))

print('TESTE')
print('Acc: %.2f (+/- %.2f)' % (np.average(cv_results['test_accuracy']), np.std(cv_results['test_accuracy'])))
print('F1: %.2f (+/- %.2f)' % (np.average(cv_results['test_f1_macro']), np.std(cv_results['test_f1_macro'])))
```

### Árvore de decisão

```python
dt = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=3)
```

```python
cv_results = cross_validate(dt, X, y, cv=10, scoring=('accuracy', 'f1_macro'))
```

```python
print('TREINO')
print('Acc: %.2f (+/- %.2f)' % (np.average(cv_results['train_accuracy']), np.std(cv_results['train_accuracy'])))
print('F1: %.2f (+/- %.2f)' % (np.average(cv_results['train_f1_macro']), np.std(cv_results['train_f1_macro'])))

print('TESTE')
print('Acc: %.2f (+/- %.2f)' % (np.average(cv_results['test_accuracy']), np.std(cv_results['test_accuracy'])))
print('F1: %.2f (+/- %.2f)' % (np.average(cv_results['test_f1_macro']), np.std(cv_results['test_f1_macro'])))
```
